```python
#importing libraries
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import urllib
import numpy as np
```

```python
#initiating variables
Brand=[]
Model=[]
Variant=[]
Year=[]
Fuel=[]
Transmission=[]
KM_driven=[]
No_of_Owners=[]
temp=[]
price=[]
```

```python
#connecting to browser and opening target page
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

url = "https://www.olx.in/moradabad_g4059318/cars_c84"
driver.get(url)
urls = []
d=0
for x in range(60):
    try:
        
        nxt_button=driver.find_element_by_xpath("//div[@class='JbJAl']")#scraping the list of buttons from the page
        nxt_button.click()
    
        time.sleep(10)
        d=d+1
    
    except:
        
        break
    
  
```

```python
#creating list of target pages
uu=driver.find_elements_by_xpath("//li[@class='EIR5N']/a")
for i in uu:
    urls.append(i.get_attribute("href"))
        
len(uu)        
```

```python
len(urls) 
```

```python
#scraping data from target pages
k=1
for URL in urls:
    
    try:
    
        
        driver.get(URL)
        datakk=driver.find_element_by_xpath("//div[@class='_3JPEe']")
        pp=datakk.text
        temp=pp.splitlines()
        print(temp)
        if temp[1]!='':
            
            Brand.append(temp[1])
        else:
            Brand.append('--')
            
            
            
        if temp[3]!='':
            
            Model.append(temp[3])
        else:
            
            Model.append('--')
            
            
        if temp[5]!='':
            
            Variant.append(temp[5])
        else:
            
            Variant.append('--')
            
        if temp[7]!='':
            Year.append(temp[7])
        else:
            
            
            Year.append('--')
        
        
        if temp[9]!='':
            
            Fuel.append(temp[9])
        else:
            
            Fuel.append('--')
            
        if temp[11]!='':
            
            Transmission.append(temp[11])
        else:
            
            Transmission.append('--')
            
        
        if temp[13]!='':
            KM_driven.append(temp[13])
        else:
            
            
            KM_driven.append('--')
            
            
        if temp[15]!='':
            
            No_of_Owners.append(temp[15])
        else:
            
            No_of_Owners.append('--')
            
            
        if (driver.find_element_by_xpath("//span[@class='_2xKfz']")).text!='':
            
            kk=(driver.find_element_by_xpath("//span[@class='_2xKfz']")).text
            kk=kk.replace(' ','')
            kk=kk.replace(',','')
            price.append(kk)
        else:
            
            
            
            price.append('--')

                
        
        
    except:
        continue
    if k==len(urls):
        break
    k=k+1

    
   
  
         
        
        

        

        
        
        
 
```

```python
#check rows in each variable
print(len(Brand))

print(len(Model))
print(len(Variant))
print(len(Year))
print(len(Fuel))

print(len(Transmission))
print(len(KM_driven))
print(len(No_of_Owners))
print(len(price))
```

```python
#creating dataset from obove variables
x=584
dataraw=pd.DataFrame({})
dataraw['Brand']=Brand[:x]
dataraw['Model']=Model[:x]
dataraw['Variant']=Variant[:x]
dataraw['Year']=Year[:x]
dataraw['Fuel']=Fuel[:x]
dataraw['Transmission']=Transmission[:x]
dataraw['KM_driven']=KM_driven[:x]
dataraw['No_of_Owners']=No_of_Owners[:x]
dataraw['price']=price[:x]
```
